#**Linear Regression using:**

#1) Normal Equation to find theta and
#2) Cost Function, then Gradient descent to find theta


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gradient descent using Random
x =2*np.random.rand(100, 1) #This is used to generate Random Numbers for x
y = 4+ 3*x + np.random.rand(100, 1) # the y = F(x) equation

#x_new = np.array([[0], [2]]) # Not sure what is for

# Give x and y the same dimension
x_b = np.c_[np.ones((100,1)), x] #Giving the variable x the same dimension as theta

# Normal Equation to find theta: 1)- One way to get Theta
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)

eta = 0.1 # learning rate
n_iteration = 1000 # number of iteration
m = 100 # number of point

theta = np.random.randn(2,1)  # Randon Initialization

 #Gradient Descent using COST FUNCTION to find theta-- for loop
for i in range(n_iteration):
  gradient = 2/m*x_b.T.dot(x_b.dot(theta) - y) # Derivative of Cost Funct%
  theta = theta - eta*gradient  # Gradient descent Formula: 2)- A way to get Theta
  y_predict = x_b.dot(theta)  # Linear Regression Prediction Formula


plt.plot(x,y, 'r.') # Plot to see how the data looks like
plt.plot(x, y_predict, 'b-')
plt.ylabel('F(theta, x)')
plt.xlabel('x(axis)')
plt.show()

```
